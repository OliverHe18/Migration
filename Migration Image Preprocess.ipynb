```python
img_path = 'test_images_2'
img_names = np.array(glob(img_path + '/*'))
print('Total number of images: ', len(img_names))

imgs = []
for img in img_names:
    imgs.append(img)

# sort 
imgs.sort()
```

```python
print(*imgs, sep='\n')
```

```python
# crop to 256*256
area = (860, 200, 1116, 456)

for i, img in enumerate(imgs, 1):
    pic = Image.open(img)
    pic_cropped = pic.crop(area)
    pic_cropped.save('test_images_2_cropped/'+ str(i) + '.png')
```
