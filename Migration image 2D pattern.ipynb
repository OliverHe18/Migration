```python
# two images diff heatmap

# simple method: diff of avg pix
# alternative: max error of each channel 

# detailed crop 
img1 = Image.open('test_images_1/30min.jpg')
img2 = Image.open('test_images_1/120min.jpg')

area = (939, 260, 972, 300)
img1_cropped = img1.crop(area)
img2_cropped = img2.crop(area)

img_1_3d = np.array(img1_cropped)
img_2_3d = np.array(img2_cropped)
plt.imshow(img_1_3d)
plt.show()
plt.imshow(img_2_3d)
plt.show()

img_1_2d = np.mean(img_1_3d, axis=2)
img_2_2d = np.mean(img_2_3d, axis=2)

diffs = img_2_2d - img_1_2d
sns.heatmap(diffs)
```

```python
# abs diff crop 
img1 = Image.open('test_images_1/30min.jpg')
img2 = Image.open('test_images_1/120min.jpg')

area = (892, 240, 1052, 320)
img1_cropped = img1.crop(area)
img2_cropped = img2.crop(area)

img1_3d = np.array(img1_cropped)
img2_3d = np.array(img2_cropped)
plt.imshow(img1_3d)
plt.show()
plt.imshow(img2_3d)
plt.show()

# Calculate the absolute difference on each channel separately
error_r = np.fabs(np.subtract(img2_3d[:,:,0], img1_3d[:,:,0]))
error_g = np.fabs(np.subtract(img2_3d[:,:,1], img1_3d[:,:,1]))
error_b = np.fabs(np.subtract(img2_3d[:,:,2], img1_3d[:,:,2]))

# Calculate the maximum error for each pixel
diff = np.maximum(np.maximum(error_r, error_g), error_b)
sns.heatmap(diff)
```
